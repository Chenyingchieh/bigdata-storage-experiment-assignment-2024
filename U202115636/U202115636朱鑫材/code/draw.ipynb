```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# cuckoo
nums = ['2','3','4','8']
percent  = [49.19, 91.78, 97.68, 99.97]
# 柱状图
plt.figure(dpi=300)
bars = plt.bar(nums, percent, label='cuckoo',width=0.5)
plt.xlabel('num')
plt.ylabel('percent')
plt.xticks(nums)

# 在每个柱子的顶部显示y轴的值
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, round(yval, 2), ha='center', va='bottom')

plt.show()
```

```python
# cuckoo
nums = ["2", "3", "4", "8"]
percent = [89.56, 95.80, 98.03, 99.78]
# 柱状图
plt.figure(dpi=300)
bars = plt.bar(nums, percent, label="cuckoo", width=0.5)
plt.xlabel("num")
plt.ylabel("percent")
plt.xticks(nums)

# 在每个柱子的顶部显示y轴的值
for bar in bars:
    yval = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        yval + 0.05,
        round(yval, 2),
        ha="center",
        va="bottom",
    )

plt.show()
```

```python
stashs = ['10', '50', '100', '200']
percent = [51.05, 52.57, 53.56, 54.75]

# 柱状图
plt.figure(dpi=300)
bars = plt.bar(stashs, percent, label="cuckoo", width=0.5)
plt.xlabel("stash")
plt.ylabel("percent")
plt.xticks(stashs)

for bar in bars:
    yval = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        yval + 0.05,
        round(yval, 2),
        ha="center",
        va="bottom",
    )

plt.show()
```

```python
blocks = ["2", "3", "4", "8"]
stashs = ["100", "200", "400", "800"]
percent = [[89.56, 89.54, 89.49, 89.44], [95.81, 95.81, 95.78, 95.77], [98.02, 97.99, 97.94, 97.87], [99.76, 99.74, 99.71, 99.62]]

# 绘制热力图
plt.figure(dpi=300)
plt.imshow(percent, cmap="Reds", interpolation="nearest")
plt.colorbar()
plt.xticks(range(len(stashs)), stashs)
plt.yticks(range(len(blocks)), blocks)
plt.xlabel("stash")
plt.ylabel("block")

# 在每个单元格上显示对应的数据
for i in range(len(blocks)):
    for j in range(len(stashs)):
        plt.text(j, i, percent[i][j], ha="center", va="center", color="black")

plt.show()
```
